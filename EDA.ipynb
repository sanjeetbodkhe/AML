```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
df = pd.read_csv(r"Telco-Customer-Churn.csv")
```

```python
#boxplot = df.boxplot(column=['tenure'])
plt.figure(figsize=(10, 5))
sns.boxplot(x='PaymentMethod', y='MonthlyCharges', data=df)
plt.title('TotalCharges by PaymentMethod')
plt.show()
```

```python
#boxplot = df.boxplot(column=['tenure'])
plt.figure(figsize=(10, 5))
sns.boxplot(x='gender', y='MonthlyCharges', data=df)
plt.title('TotalCharges by PaymentMethod')
plt.show()
```

```python
#df.plot.scatter(x = 'tenure', y = 'TotalCharges', s = '1');
plt.figure(figsize=(12, 6))
scatter = sns.scatterplot(x='tenure', y='TotalCharges', data=df)
scatter.set_title('Tenure vs Total Charges')
scatter.set_xlabel('Tenure')
scatter.set_ylabel('Total Charges')
plt.yscale('log')
plt.show()
```

```python
plt.figure(figsize=(7, 7))
stats.probplot(df['MonthlyCharges'], dist="norm", plot=plt)
plt.title('QQ Plot of Monthly Charges')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Ordered Values')
plt.grid()
plt.show()
```

```python
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.gofplots import qqplot_2samples
```

```python
x = df[df['gender']=='Male']['MonthlyCharges']
y = df[df['gender']=='Female']['MonthlyCharges']
pp_x = sm.ProbPlot(x)
pp_y = sm.ProbPlot(y)
qqplot_2samples(pp_x, pp_y,xlabel="Male Monthly Charges",ylabel="Female Monthly Charges", line='45')
plt.show()
```

```python
categorical_features = ['gender','SeniorCitizen','Partner','Dependents','PhoneService','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','Contract','PaperlessBilling','PaymentMethod']

for feature in categorical_features:
    plt.figure(figsize=(8, 5))
    sns.countplot(x=feature, hue='Churn', data=df)
    plt.title(f'{feature} Distribution by Churn')
    plt.show()
```

```python
plt.figure(figsize=(20, 5))
sns.countplot(x='tenure', hue='Churn', data=df)
plt.title('Tenure Distribution by Churn')
plt.show()
```
